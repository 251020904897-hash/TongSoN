cho tổng chẵn bằng 0, tổng lẻ bằng 0,
nhập một bảng nguyên gồm n phần tử,
tính tổng các phần tử số chẵn và số lẻ

```python
tong_chan = 0
tong_le = 0
n = int(input("Nhập số phần tử n: "))
for i in range(n):
  x = int(input(f"Nhập phần tử thứ {i+1}: "))
  if x % 2 == 0:
    tong_chan += x
  else:
    tong_le += x
print("Tổng các số chẵn =", tong_chan)
print("Tổng các số lẻ   =", tong_le)



```
